## 7.11 Plots

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas
```

```python
df_0 = pandas.read_csv('../RawData1_6/A05D0R_1.csv', skiprows=26)
df_45 = pandas.read_csv('../RawData1_6/A05D45R_1.csv', skiprows=26)
df_90 = pandas.read_csv('../RawData1_6/A05D90R_1.csv', skiprows=26)
```

```python
area_0 = 23.29E-3 * 2.43E-3
area_45 = 24.59E-3 * 2.74E-3
area_90 = 24.39E-3 * 2.8E-3
```

```python
stress_0 = (df_0['Force'][1:].to_numpy().astype(float) * 1000) / (area_0 * 1.0E6)
strain_0 = df_0['Tensile strain (Strain 1)'][1:].to_numpy().astype(float)
```

```python
plt.plot(strain_0, stress_0)
plt.ylabel('Stress [MPa]')
plt.xlabel('Strain [mm/mm]')
plt.title('0 degree: Stress vs Strain')
plt.xlim(0, 0.005)
plt.ylim(0, 375)
plt.grid()
plt.show()
```

```python
stress_45 = (df_45['Force'][1:].to_numpy().astype(float) * 1000) / (area_45 * 1.0E6)
strain_45 = df_45['Tensile strain (Strain 1)'][1:].to_numpy().astype(float)
plt.plot(strain_45, stress_45)
plt.ylabel('Stress [MPa]')
plt.xlabel('Strain [mm/mm]')
plt.title('45 degree: Stress vs Strain')
plt.grid()
plt.show()
```

```python
stress_90 = (df_90['Force'][1:].to_numpy().astype(float) * 1000) / (area_90 * 1.0E6)
strain_90 = df_90['Tensile strain (Strain 1)'][1:].to_numpy().astype(float)
plt.plot(strain_90, stress_90)
plt.ylabel('Stress [MPa]')
plt.xlabel('Strain [mm/mm]')
plt.title('90 degree: Stress vs Strain')
plt.grid()
plt.show()
```
